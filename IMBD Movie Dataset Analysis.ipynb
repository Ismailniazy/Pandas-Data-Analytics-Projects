#### CLASSIFY MOVIES BASED ON RATING[EXCELLENT,GOOD,AVERAGE & BAD]

```python
def rating(rating):
    if rating >=8.0:
        return 'Excellent'
    elif rating >=6.0:
        return 'Good'
    elif rating >=4.0:
        return 'Average'
    else:
        return 'Bad'
```

```python
DF['Rating_cat'] = DF['Rating'].apply(rating)
```

```python
DF.head(2)
```

#### FIND UNIQUE VALUES FROM GENRE

```python
list1=[]
for value in DF['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique_list=[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
        
```

```python
unique_list
```

```python
len(unique_list)
```

```python
print('NUMBER OF UNIQUE GENRES =',len(unique_list))
```
